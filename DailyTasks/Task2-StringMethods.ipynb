**Q3**: What is the difference between string.find() and string.index() ? Explain with examples

```python
# Both string.find() and string.index() are used to search for a substring in a string 
# but find() gives -1 if it doesn't find the substring whereas index() will throw an error
s1 = "InnomaticsResearchLabs"
print(s1.find("i"))
print(s1.index("i"))
```

```python
#this is the difference
print(s1.find("Z"))
print(s1.index("Z"))
```

**Q6**: Check if the string s (s = "madam") is a palindrome and assign the result (True/False) to the variable is_palindrome.

```python
s = "madam"
rev=s[::-1]
is_palindrome = s==rev
print(is_palindrome)
```
